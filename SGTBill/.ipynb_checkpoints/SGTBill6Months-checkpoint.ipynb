```python
# If module not found error is raised try the workaround below
# import sys
# print(sys.executable) 
# copy the output of the print statement and install the missing module
# Example: /opt/homebrew/Cellar/jupyterlab/4.2.1/libexec/bin/python -m pip install pandas
import pandas as pd
import matplotlib.pyplot as plt

# We read the CSV exported from the MAS website 
# (https://eservices.mas.gov.sg/statistics/fdanet/BondTreasuryBillsCMTBsAuctions.aspx) 
# and during reading skips the first row as it is basically a title for the file.
# TODO: Look into extracting/importing the data directly from the MAS website APIs
tbill_df = pd.read_csv('SGS Bonds and T-bills & CMTBs Auction Result Data.csv',skiprows=[0])

# Cleanup by dropping irrelevant columns for our analysis
tbill_df.drop(axis=1,columns=['Product Type','Coupon Rate (%)'],inplace=True)

# Convert some of the column to correct data types to ease future analysis
# TODO: Need to check if there is a more optimized way of doing this during the reading 
#       of the data from file.
tbill_df["Term to Maturity at Auction"]= pd.to_numeric(tbill_df["Term to Maturity at Auction"])
tbill_df["Auction Amount (S$M)"]= pd.to_numeric(tbill_df["Auction Amount (S$M)"])
tbill_df["Auction Date"]= pd.to_datetime(tbill_df["Auction Date"],dayfirst = True)
tbill_df["Issue Date"]= pd.to_datetime(tbill_df["Issue Date"],dayfirst = True)
tbill_df["Maturity Date"]= pd.to_datetime(tbill_df["Maturity Date"],dayfirst = True)
tbill_df["Bid-to-Cover Ratio"]= pd.to_numeric(tbill_df["Bid-to-Cover Ratio"])
tbill_df["Cut-off Yield (%)"]= pd.to_numeric(tbill_df["Cut-off Yield (%)"])
tbill_df["Cut-off Price"]= pd.to_numeric(tbill_df["Cut-off Price"])
tbill_df["Median Yield (%)"]= pd.to_numeric(tbill_df["Median Yield (%)"])
tbill_df["Median Price"]= pd.to_numeric(tbill_df["Median Price"])
tbill_df["Average Yield (%)"]= pd.to_numeric(tbill_df["Average Yield (%)"])
tbill_df["Average Price"]= pd.to_numeric(tbill_df["Average Price"])

# Display the data frame to assist in the analysis during development
tbill_df
```

```python
# Source
# STI: https://www.investing.com/indices/singapore-straits-time-historical-data

# Import the Straits Times Index data for the same historical period as the SG T-Bill
sti_df = pd.read_csv("FTSE Straits Times Singapore Historical Data.csv")

# Convert some of the column to correct data types to ease future analysis
# TODO: Need to check if there is a more optimized way of doing this during the reading 
#       of the data from file.
sti_df["Date"]= pd.to_datetime(sti_df["Date"])
sti_df["Price"] = sti_df["Price"].str.replace(",", "").astype(float)
sti_df["Open"] = sti_df["Open"].str.replace(",", "").astype(float)
sti_df["High"] = sti_df["High"].str.replace(",", "").astype(float)
sti_df["Low"] = sti_df["Low"].str.replace(",", "").astype(float)
sti_df["Change %"]= sti_df["Change %"].str.replace("%","").astype(float)

# Convert string to scientific notation and eventually to float to ease future analysis
sti_df["Vol."] = sti_df["Vol."].replace({"K": "e+3","M": "e+6","B": "e+9"},regex = True).astype(float)

# Display the data frame to assist in the analysis during development
sti_df
```

```python
# Source
# SGX: https://www.investing.com/equities/singapore-exchange-historical-data?utm_source=google&utm_medium=cpc&utm_campaign=18959642585&utm_content=635743734396&utm_term=dsa-1546555491534_&GL_Ad_ID=635743734396&GL_Campaign_ID=18959642585&ISP=1&npl=1&af_adset_id=144433385435&gad_source=1&gclid=CjwKCAjw-O6zBhASEiwAOHeGxXEX4XQgWW7BKunqo5mJ8wg3n4dGFPPiOApxLUSjoXzm77gRxk2qoRoCsEkQAvD_BwE

# Import the SGX Price data for the same historical period as the SG T-Bill
sgx_df = pd.read_csv("SGXL Historical Data.csv")

# Convert some of the column to correct data types to ease future analysis
# TODO: Need to check if there is a more optimized way of doing this during the reading 
#       of the data from file.
sgx_df["Date"]= pd.to_datetime(sgx_df["Date"])
sgx_df["Price"]= pd.to_numeric(sgx_df["Price"])
sgx_df["Open"]= pd.to_numeric(sgx_df["Open"])
sgx_df["High"]= pd.to_numeric(sgx_df["High"])
sgx_df["Low"]= pd.to_numeric(sgx_df["Low"])
sgx_df["Change %"]= sgx_df["Change %"].str.replace("%","").astype(float)

# Convert string to scientific notation and eventually to float to ease future analysis
sgx_df["Vol."] = sgx_df["Vol."].replace({"K": "e+3","M": "e+6","B": "e+9"}, regex=True).astype(float)

# Display the data frame to assist in the analysis during development
sgx_df
```

```python
# Source https://fred.stlouisfed.org/series/FEDFUNDS

# Import the USA Federal Funding Rate data for the same historical period as the SG T-Bill
usfed_df = pd.read_csv("FEDFUNDS.csv")

# Convert some of the column to correct data types to ease future analysis
# TODO: Need to check if there is a more optimized way of doing this during the reading 
#       of the data from file.
usfed_df["DATE"]= pd.to_datetime(usfed_df["DATE"])
usfed_df["FEDFUNDS"]= pd.to_numeric(usfed_df["FEDFUNDS"])

usfed_df
```

```python
fig,ax = plt.subplots(4)
fig.set_size_inches(20.0,20.0)
plt.subplots_adjust(hspace = 0.28)

ax[0].set_xlabel("Year")
ax[0].set_ylabel("Cut-off Yield (%)")
ax[0].set_title("Singapore 6 months T-Bill Yields (Cut-off Yield)")
ax[0].plot(tbill_df["Issue Date"],tbill_df["Cut-off Yield (%)"],'g')
#ax[0].plot(df["Issue Date"],df["Median Yield (%)"],'b')

ax[2].set_xlabel("Year")
ax[2].set_ylabel("STI Price (SGD$)")
ax[2].set_title("Singapore Straits Times Index")
ax[2].plot(sti_df["Date"],sti_df["High"],"b")

ax[3].set_xlabel("Year")
ax[3].set_ylabel("SGX Price (SGD$)")
ax[3].set_title("SGX Price")
ax[3].plot(sgx_df["Date"],sgx_df["High"],"r")

ax[1].set_xlabel("Year")
ax[1].set_ylabel("Fed Fund Rate (%)")
ax[1].set_title("USA Federal Funding Rate")
ax[1].plot(usfed_df["DATE"],usfed_df["FEDFUNDS"],"y")

plt.show()
```
